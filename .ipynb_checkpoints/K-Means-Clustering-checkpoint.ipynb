9. เลือกฟีเจอร์ STG กับ PEG สร้าง K-Means ให้ K=2

```python
from sklearn.cluster import KMeans
```

```python
X = df[['STG', 'PEG']]
X
```

```python
kmeans = KMeans(n_clusters = 2, random_state = 111)
```

```python
kmeans_label = kmeans.fit_predict(X)
kmeans_label
```

10. หาค่า cluster coordinates, inertia และ iteration ของ K-Means


```python
# Coordinates of cluster centers.
kmeans.cluster_centers_
```

```python
# Sum of squared distance of sample to their closest cluster center. float
kmeans.inertia_
```

```python
# Number of iterations run. int
kmeans.n_iter_
```

11. สร้าง Clustering Visualization

```python
x1 = X[kmeans_label == 0].iloc[:,0]
y1 = X[kmeans_label == 0].iloc[:,1]
x2 = X[kmeans_label == 1].iloc[:,0]
y2 = X[kmeans_label == 1].iloc[:,1]
```

```python
x1
```

```python
df2 = df.copy()
df2['cluster'] = kmeans_label
df2
```

```python
fig = plt.figure(figsize=(8,5))
sns.scatterplot(data=df2, x='STG', y='PEG', 
                hue='cluster', s=50)
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], 
                y=kmeans.cluster_centers_[:, 1],
                color='g', s=150, marker='X')
plt.legend(['1', '2'], title='Cluster',loc='lower right')
```

12. ทำ Elbow Method หาค่า K ที่เหมาะสมที่สุด และทำ Elbow Method Visualization 

```python
# find the best K
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, random_state = 11)
    kmeans.fit(X)
    # remember that inrtia is wcss per each K
    wcss.append(kmeans.inertia_)

fig = plt.figure(figsize=(12,8))
sns.lineplot(x=range(1,11), y=wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
```

14.  สร้าง Clustering Visualization ของ K-Means

15.  เลือกฟีเจอร์ใดก็ได้ 2 ฟีเจอร์ที่ไม่ซ้ำเดิม สร้าง K-Means ให้ K=2

```python
df.head(10)
```

```python
df['UNS'].unique()
```

```python
df_final = df.copy()
df_final.replace({'very_low':1, 'Low':2, 
                  'Middle':3, 'High':4}, inplace=True)
```

```python
df_final
```

```python
df_final.corr()
```

```python
from sklearn.preprocessing import MinMaxScaler
#Scaling
#Normalization
min_max_scaler = MinMaxScaler()

arr_minmax1 = min_max_scaler.fit_transform(df_final)
#arr_minmax2 = min_max_scaler.transform(X_test)

X_n_train = pd.DataFrame(arr_minmax1, columns=df_final.columns)
#X_n_test = pd.DataFrame(arr_minmax2, columns=X_train.columns)
```

```python
# PCA 2 components
from sklearn.decomposition import PCA 

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_n_train)
#X_test_pca = pca.transform(X_n_test)
```

```python
X_train_pca
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=df_final.columns)
df_comp
```

```python
# K-means
kmeans = KMeans(n_clusters = 2, random_state = 111)
kmeans_label = kmeans.fit_predict(X_train_pca)
kmeans_label
```

16. หาค่า cluster coordinates ของ K-Means ของฟีเจอร์ใหม่

```python
# Coordinates of cluster centers.
kmeans.cluster_centers_
```

17. หาค่า inertia และ iteration ของ K-Means ของฟีเจอร์ใหม่

```python
# Sum of squared distance of sample to their closest cluster center. float
kmeans.inertia_
```

```python
# Number of iterations run. int
kmeans.n_iter_
```

18. สร้าง Clustering Visualization ของฟีเจอร์ใหม่

```python
plt.figure(figsize=(8,6))
plt.scatter(X_train_pca[:,0], X_train_pca[:,1], c=kmeans_label)
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], 
                y=kmeans.cluster_centers_[:, 1],
                color='b', s=200, marker='X')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.grid()
```


19. ทำ Elbow Method หาค่า K ที่เหมาะสมที่สุด ของฟีเจอร์ใหม่ และทำ Elbow Method Visualization

```python
# find the best K
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, random_state = 11)
    kmeans.fit(X_train_pca)
    # remember that inrtia is wcss per each K
    wcss.append(kmeans.inertia_)

fig = plt.figure(figsize=(12,8))
sns.lineplot(x=range(1,11), y=wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
```
